```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
import seaborn as sns
#importing matplotlib / pandas
```

```python
df = pd.read_csv("percentage_of_visits_ili_2.csv")
df
```

```python
#This data is pretty clean and does not need much cleaning
```

```python
corr = df.corr() 
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(10, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);

plt.savefig("correlation_ILI")
```

```python
df.corr()
```

The inference: 
1. With increase in MMRW week there was decrease in cases amongst people who are above 65 years of age;
2. Positive correlation between unweighted ILI and weighted ILI 
3. With increase in MMRW week there is decrease in cases recorded amongst people who are b/w 50-64 years;
4. Shockingly there is a correlation between Total Patients and Age 0-4, this can indicate that more number of contribution to total patients is from patients between the Age 0 and 4
5. There is a difference between Total ILI and Total Patients; Age 5-24 have been contributing more for Total Influenza like illness count.
